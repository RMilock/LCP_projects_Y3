## POINT 4

 Study of the visitors' fluxes: let's assume to be asked to provide indications of how to invest resources to improve the mobility toward Padova. Defined the three main directions of visitors and commuter getting to Padova by the Italian highways (sud: A13 toward Bologna-Roma / west: A4 toward Milano-Torino / north-east: A4 toward Venice-Trieste), evaluate which of the three directions has to be prioritized:
 1. Consider a simplified case involving only the mid-range mobility, based on the number of visitors/commuters from the nearby regions only
 * Consider the provinces located on the three directions that are mostly contributing to the flow of weekend visitors and working daily commuters by performing a more detailed study of the fluxes based on the day of the week
 * _[OPEN]_ Use the data available to provide what you believe is the best possible answer 

```python
def codici(codici_istat_provincia, capoluogo):
    cod_reg=codici_istat_provincia[codici_istat_provincia['PROVINCIA'] == capoluogo]['COD_REG'].values[0] #codice comune di Padova
    province=codici_istat_provincia[codici_istat_provincia['COD_REG'] == cod_reg]
    cod_all_prov=province['COD_PRO'].values[:]
    name_all_prov=province['PROVINCIA'].values[:]
    return cod_reg, province, cod_all_prov, name_all_prov



#Veneto 
cod_reg_veneto, info_veneto, cod_all_prov_veneto, name_prov_veneto=codici(codici_istat_provincia, 'Venezia')
#print(prov_veneto)
print("Region code of  Veneto: ",cod_reg_veneto)
print("Provinces of Veneto: ", name_prov_veneto)
print("Province codes of Veneto: ", cod_all_prov_veneto)
print("Code of Padua ", codici_istat_provincia[codici_istat_provincia['PROVINCIA']=='Padova']['COD_PRO'].values[0])
print("\n")

#Lombardia
cod_reg_lombardia, info_lombardia, cod_all_prov_lombardia, name_prov_lombardia=codici(codici_istat_provincia, 'Milano')
#print(province_lombardia)
print("Region code of  Lombardia: ",cod_reg_lombardia)
print("Provinces of Lombardia: ", name_prov_lombardia)
print("Province codes of  Lombardia: ",cod_all_prov_lombardia)
print("\n")


#Emilia Romagna
cod_reg_er, info, cod_all_prov_er, name_prov_er=codici(codici_istat_provincia, 'Bologna')
#print(province_veneto)
print("Region code of Emilia Romagna: ",cod_reg_er)
print("Provinces of Emilia Romagna: ", name_prov_er)
print("Province codes of dell' Emilia Romagna: ", cod_all_prov_er)
print("\n")

#Friuli Venezia Giulia
cod_reg_friuli, info_friuli, cod_all_prov_friuli, name_prov_friuli=codici(codici_istat_provincia, 'Trieste')
print(info_friuli)
print("Region code of Friuli Venezia Giulia: ", cod_reg_friuli)
print("Provinces of Friuli Venezia Giulia: ", name_prov_friuli)
print("Province codes of  Friuli Venezia Giulia: ", cod_all_prov_friuli)
print("\n")


#POINT 4.1 : study of the fluxes of the nearby regions
data_day_od


def flow_regions(data_day_od, cod_all_prov):
    flow_sum=data_day_od.groupby([data_day_od['COD_PRO'].isin(cod_all_prov)])['FLOW'].sum()
    return flow_sum[1] #true is the second one 


#Veneto #per averlo per ora non utile ai fine dell'analisi
flow_veneto=flow_regions(data_day_od, cod_all_prov_veneto)
print("Flow Veneto: ",flow_veneto)

#Lombardia
flow_lombardia=flow_regions(data_day_od, cod_all_prov_lombardia)
print("Flow Lombardia: ",flow_lombardia)

#Emilia Romagna
flow_er=flow_regions(data_day_od, cod_all_prov_er)
print("Flow Emilia Romagna: ",flow_er)

#Friuli Venezia Giulia
flow_friuli=flow_regions(data_day_od, cod_all_prov_friuli)
print("Flow Friuli Venezia Giulia: ",flow_friuli)


#POINT 4.2 : Flow of weekend visitors and working daily commuters:
#flux based on the day of the week, mostly contributing provinces located on the three directions 

#DATASET REGIONI DI INTERESSE

def data_province(data_day_od, cod_all_prov):
    data_province=data_day_od[data_day_od['COD_PRO'].isin(cod_all_prov)]
    return data_province


data_prov_lombardia=data_province(data_day_od, cod_all_prov_lombardia)
data_prov_er=data_province(data_day_od, cod_all_prov_er)
data_prov_friuli=data_province(data_day_od, cod_all_prov_friuli)
data_prov_veneto=data_province(data_day_od, cod_all_prov_veneto) #TOGLIERE PADOVA



#SUM FLUX PER PROVINCIA: SELEIONE DELLE PROVINCE CHE CONTRIBUISCONO MAGGIORMENTE 

def flow_province(data_province):
    flow_province=data_province.groupby('COD_PRO').FLOW.sum()
    flow_sorted=flow_province.sort_values(ascending=False)
    #flow_selected=flow_sorted.loc[flow_sorted[:] >= 1e4]
    flow_selected= flow_sorted.head(n=3).reset_index()
    cod_pro_selected=flow_selected['COD_PRO'].values[:]
    return flow_selected, cod_pro_selected


flow_province_lombardia, cod_pro_selected_lombardia=flow_province(data_prov_lombardia)
print("Lombardia \n",flow_province_lombardia) 
print("Code selected provinces of Lombardia \n",cod_pro_selected_lombardia) 
print("\n")

flow_province_er, cod_pro_selected_er=flow_province(data_prov_er)
print("Emilia Romagna \n",flow_province_er)  #codici provincia non chiamabili
print("Code selected provinces of Emilia Romagna \n",cod_pro_selected_er) 
print("\n")


flow_province_friuli, cod_pro_selected_friuli=flow_province(data_prov_friuli)
print("Friuli Venezia Giulia \n",flow_province_friuli) 
print("Code selected provinces of Friuli Venezia Giulia \n",cod_pro_selected_friuli) 
print("\n")

#####flow_province_veneto=flow_province(data_prov_veneto)
#print("Veneto \n",flow_province_veneto) 


#Sum flux per i giorni della settimana in base alle province scelte
def flow_days(data_province, cod_pro_selected):
    data_province_selected=data_province[data_day_od['COD_PRO'].isin(cod_pro_selected)] #PROBLEMI!!!
    print(data_province_selected)
    flow_days=data_province_selected.groupby(['DOW', 'COD_PRO']).FLOW.sum().reset_index()
    #flow_sorted=flow_days['FLOW'].sort_values(ascending=False)
    #flow_selected=flow_sorted.loc[flow_sorted[:] >= 1e4]
    #flow_selected= flow_sorted.head(n=3)
    return flow_days

flow_days_lombardia=flow_days(data_prov_lombardia,cod_pro_selected_lombardia)
print(flow_days_lombardia)
print(flow_days_lombardia.sort_values(by='FLOW',ascending=False))
print(flow_days_lombardia.sort_values(by='COD_PRO',ascending=False))
print("\n")

#ordine per i giorni della settimana
cats = ['Lunedì', 'Martedì', 'Mercoledì', 'Giovedì', 'Venerdì', 'Sabato', 'Domenica']
flow_days_lombardia['DOW'] = pd.Categorical(flow_days_lombardia['DOW'], categories=cats, ordered=True)
flow_days_lombardia_days = flow_days_lombardia.sort_values(['DOW','COD_PRO'])

print("laa \n",flow_days_lombardia_days)


#nei weekends 
```
